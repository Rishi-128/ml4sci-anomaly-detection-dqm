This synthetic dataset simulates detector measurements.
Normal samples represent stable detector conditions,
while anomalous samples simulate abnormal behavior.
The goal is to train a model only on normal data
and detect anomalies via reconstruction error.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Normal data (simulate healthy detector behavior)
normal_data = np.random.normal(loc=0.0, scale=1.0, size=(10000, 10))

# Anomalous data (simulate detector issues)
anomalies = np.random.normal(loc=3.0, scale=1.5, size=(500, 10))

data = np.vstack([normal_data, anomalies])
labels = np.hstack([np.zeros(len(normal_data)), np.ones(len(anomalies))])

print("Data shape:", data.shape)
print("Anomalies:", labels.sum())
```

```python
plt.figure(figsize=(6,4))
plt.hist(data[:,0], bins=50)
plt.title("Feature 0 distribution")
plt.show()
```
